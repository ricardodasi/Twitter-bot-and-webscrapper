```python
#Script for scraping scryfall websites for all magic the gathering cards

#import packages 

import requests 

from bs4 import BeautifulSoup

from pymongo import MongoClient

from mongoengine import * 
```

```python

#establishing connections

connect('mongoengine_test', host='localhost', port=27017)

#setting up connection to local client

client = MongoClient()

#setting up connection to magic_cards database

db = client.magic_cards

#setting up connection to cards_collection collection

cards_collection = db.cards_collection


#creating class for the cards

#creating a class for my cards objects 


class Card(Document):
    title = StringField(required=True)
    mana_cost = StringField(required=True)
    edition = StringField(required=True)
    effect_text = StringField(required=True)
    flavour_text = StringField(required=True)
    type_of_card = StringField(required=True)
    artist = StringField(required=True)
    image = StringField(required=True)
    price = StringField(required=True)
    
    
    def save(self):
        file = {'title':self.title,
                'mana_cost':self.mana_cost,
                'edition':self.edition,
                'effect_text':self.effect_text,
                'flavour_text':self.flavour_text,
                'type of card':self.type_of_card,
                'artist':self.artist,
                'image':self.image,
                'price':self.price
               }
        
        cards_collection.insert_one(file)
```

```python
#Final script for scraping the whole data base of scryfall


#requesting all the editions websites 

page = requests.get('https://scryfall.com/sets/')


#creating soup object  and parsing the html


soup = BeautifulSoup(page.content, 'html.parser')


#we are looking for the class 'card-grid-item-card' so we select those and create a list of links to scrape 

editions_library = []


for i in range(len(soup.findAll(class_="flexbox"))): 
    editions_library.append(soup.findAll(class_="flexbox")[i].find('a').get('href'))
    print(editions_library[i])
    
    
#We remove atypical links that do not correspond to editions but particular formats    
    
for i in editions_library:
    if 'https' not in i:
        editions_library.remove(i)


```

```python

#Now we iterate over this list of links with the previous code to get all the cards        


for edition in editions_library:
    
    page = requests.get(edition)


    #creating soup object  and parsing the html


    main_soup = BeautifulSoup(page.content, 'html.parser')


    #we are looking for the class 'card-grid-item-card' so we select those and create a list of links to scrape 

    main_library =[]
    
    cards_with_errors = []

    for i in range(len(main_soup.find_all(class_="card-grid-item-card"))):
        if i < len(main_soup.find_all(class_="card-grid-item-card")):
            main_library.append(main_soup.find_all(class_="card-grid-item-card")[i].get('href'))  
        else:
            break
            
        
    #now we get each of the cards uploaded to the data base

    for card in main_library:

        try:

            page =  requests.get(card)

            inside_soup = BeautifulSoup(page.content, 'html.parser')      


            #getting the items that are variable from card to card (flavour and effect text, prices, mana cost):


            #getting card effect text 

            if not inside_soup.find_all(class_="card-text-oracle"):
                effect_text = 'empty'
                

            else: 
                effect_text = inside_soup.find_all(class_="card-text-oracle")[0].get_text().strip()
                

            #getting card flavour text 

            if not inside_soup.find_all(class_="card-text-flavor"):
                flavour_text = 'empty'
                

            else:               
                flavour_text = inside_soup.find_all(class_="card-text-flavor")[0].get_text().strip()
                

            #getting price

            for i in range(len(inside_soup.find_all(class_='current'))):
                try:
                    price = inside_soup.find_all(class_='current')[i].find(class_='currency-usd').get_text()
                except:
                    pass
                
            #getting mana cost 
            
            if not inside_soup.find_all(class_='card-text-mana-cost'):
                mana_cost = 'no cost'
            else:
                mana_cost = inside_soup.find_all(class_='card-text-mana-cost')[0].get_text()
                
            

            #store the card in a card object and upload it to the data base in json format, moves on to the next card

            current_card = Card(title = inside_soup.find_all('title')[0].get_text().split('·')[0].strip(),
                                mana_cost = mana_cost,
                                edition = inside_soup.find_all('title')[0].get_text().split('·')[1].strip(),
                                effect_text =  effect_text,
                                flavour_text = flavour_text,
                                type_of_card = inside_soup.find_all(class_="card-text-type-line")[0].get_text().strip(),
                                artist = inside_soup.findAll(class_="card-text-artist")[0].get_text().split('by')[1].strip(),
                                image = inside_soup.findAll('img')[0].get('src'),
                                price = price 
                               )

            current_card.save()
            print(current_card.title)



        #raising exception for some cards with issues to check them later.

        except:         
                #we append the title of the failed card to the list
                error_card_name = inside_soup.find_all('title')[0].get_text().split('·')[0].strip()
                cards_with_errors.append(error_card_name)
                print('error with {}'.format(error_card_name))
                pass
    
    

```
